#### 1.2.2 팀 경력 합쳐주기

```python
temp_3 = data_2[~(((data_2.team.isnull()) & (data_2.team_now.isnull())) | ((data_2.team.isnull()) & (data_2.team_now == '-')))]
temp_3
```

```python
temp_3.team.iloc[0]
```

```python

def club_career_remover(index_num):
    temp_4 = temp_3.team.iloc[index_num]
    temp_4 = temp_4.replace('임대','').replace('~','')
    temp_4 = temp_4.split('\n')
    temp_4 = list(filter(None, temp_4))

    temp_5 = []
    for x in range(len(temp_4)):
        if x % 2 == 1:
            temp_5.append(temp_4[x])
            temp_5 = list(set(temp_5))
            
    return temp_5

```

```python
temp_3.team.iloc[1222]
```

```python
# 먼저 NaN 값 부터 처리후 밑 코드 돌리기!

for index in range(len(temp_3)):
    temp_a = club_career_remover(index)
    temp_3.team.iloc[index] = temp_a
    
temp_3
```
