##### 1.2.2.2 경력 정리

```python
# 선수 팀 커리어 정리 함수
def club_career_remover(index_num):
    temp_5 = temp_4.team.iloc[index_num]
    temp_5 = temp_5.replace('임대','').replace('~','')
    temp_5 = temp_5.split('\n')
    temp_5 = list(filter(None, temp_5))

    temp_6 = []
    temp_7 = ''
    for x in range(len(temp_5)):
        if temp_5[0].startswith('1') or temp_5[0].startswith('2'):
            if x % 2 == 1:
                temp_6.append(temp_5[x])
        else:
            if x % 2 == 0:
                temp_6.append(temp_5[x])
                
        
    if temp_6 == ['-']:
        temp_6 = [temp_4.team_now.iloc[index_num]]
        
    elif temp_4.team_now.iloc[index_num] == '-':
        temp_6 = temp_6
    else:
        temp_6 = temp_6 + [temp_4.team_now.iloc[index_num]]
        
    temp_6 = list(set(temp_6))
    temp_7 = ' / '.join(s for s in temp_6)
            
    return temp_7
```

```python
# 수정
temp_4['club_career'] = '-'

for index in range(len(temp_4)):
    temp_a = club_career_remover(index)
    temp_4.club_career.iloc[index] = temp_a
    
temp_4
```

### 2.1.1 국가 pk

```python
nations = pd.DataFrame(data = list(data.etc_nation.value_counts().index), columns = ['nation'])
nations
```

```python
nation_pk = []

for num in range(1,len(nations)+1):
    if num < 10:
        nation_pk.append(f'00{num}')
    elif num < 100:
        nation_pk.append(f'0{num}')
    elif num >= 100:
        nation_pk.append(f'{num}')

nations['nation_pk'] = nation_pk
nations
```

### 2.1.2 클럽 pk

```python
temp_club_list = []

for num in range(len(data)):
    temp_club_list.append(data.club_career.iloc[num].split(' / '))
    
club_list = sum(temp_club_list,[])
club_list
```

```python
club_list = list(set(club_list))
club_list
```

```python
clubs = pd.DataFrame(data = club_list, columns = ['club'])
clubs
```

```python
club_pk = []

for num in range(1,len(clubs)+1):
    if num < 10:
        club_pk.append(f'000{num}')
    elif num < 100:
        club_pk.append(f'00{num}')
    elif num < 1000:
        club_pk.append(f'0{num}')
    elif num >= 1000:
        club_pk.append(f'{num}')

clubs['club_pk'] = club_pk
clubs
```

### 2.1.3 시즌 pk

### 2.2 국가 & 클럽 pk 추출

```python
# data.drop(['Unnamed: 0'], axis=1, inplace=True)
data.tail(3)
```

```python
data['nation_pk'] = '-'
data['club_pk'] = '-'


for x in tqdm(nations.nation): # x = '잉글랜드'
    for num in list(data[data['etc_nation'].str.contains(x)].index): # '잉글랜드'를 포함하는 df의 index 
         data.nation_pk.iloc[num] = data.nation_pk.iloc[num].replace('-', ''.join(nations[nations.nation == x].nation_pk.values))


for x in tqdm(clubs.club): # x = '바르셀로나'
    for num in list(data[data['club_career'].str.contains(x)].index): # '잉글랜드'를 포함하는 df의 index 
        if data.club_pk.iloc[num] == '-':
            data.club_pk.iloc[num] = data.club_pk.iloc[num].replace('-', ''.join(clubs[clubs.club == x].club_pk.values))
        else:
            data.club_pk.iloc[num] = data.club_pk.iloc[num]+','+(''.join(clubs[clubs.club == x].club_pk.values))


data
        
```
